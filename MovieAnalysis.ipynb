```python
# The dataset has 3 different spreadsheets, the task here is to join them
# join the spreadsheets
df0 = pd.read_excel('movies.xls', sheet_name=0, index_col=0)
df1 = pd.read_excel('movies.xls', sheet_name=1, index_col=0)
df2 = pd.read_excel('movies.xls', sheet_name=2, index_col=0)
all_movies = pd.concat([df0,df1,df2])
# show total numbers of rows and columns
all_movies.shape
```

```python
#Now that we have joined the spreadsheets, let us see the first five rows of all movies
all_movies.head()
```

```python
# another method of joining sheets is by using for loop
xlsx = pd.ExcelFile('movies.xls')
all_sheets = []
for sheet in xlsx.sheet_names:
    all_sheets.append(xlsx.parse(sheet))
all_movies = pd.concat(all_sheets) 
all_movies.head()
```

```python
# let us see the summary statistics
all_movies.describe()
```

```python
#we want to know the movies that earned more
gross_sort = all_movies.sort_values(['Gross Earnings'], ascending=False)
gross_sort.head()
```

```python
# also, let us see the movies with more budget
gross_sort = all_movies.sort_values(['Budget'], ascending=False)
gross_sort.head()
```

```python
gross_sort['Gross Earnings'].head(10)
```

```python
# movies that was rated high by IMDB
rating_sort = all_movies.sort_values(['IMDB Score'], ascending=False)
rating_sort.head(2)
```

```python
# let's visualize first fifteen movies with highest ratings
rating_sort['IMDB Score'].head(15).plot(kind='bar')
```
